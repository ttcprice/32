```python
class MainFlow(Workflow):

    def id(self):
        return "CollectReportFlow"

    def version(self):
        return 1

    def build(self, builder: WorkflowBuilder):
        builder\
            .start_with(StartFlow)\
                .then(SendNotification)\
                .then(GetReport20)\
                .then(GetReport22)\
                .then(GetReport3)\
                .then(GetReportCastody)\
                .then(GetReport2)\
                .then(GetReportSK)\
                .then(RunMainPrudikiMacro)\
                .then(RunSravnMacro)\
                .then(ParseAndSend)\
                .then(FinishFlow)


host = configure_workflow_host()
host.start()               

try:
    host.register_workflow(CollectReportFlow())
    log.info('******** Запуск робота ********')
    wid = host.start_workflow("MainFlow", 1, None)
    wait_until(250, 5, lambda: host._persistence_service.get_workflow_instance(wid).status, value=2)
except Exception as e:
    log.info(str(e))
finally:
    host.stop()
```
